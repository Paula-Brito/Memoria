```python
ruta_level01='/home/paula/Documentos/Memoria/HySea/UTFSM-IniciacionTsunami-20200505T151943Z-001/UTFSM-IniciacionTsunami/ejemplo_chile/Escenarios_corridos/Escenarios/res_level01_sce400/'
ruta = '/home/paula/Documentos/Memoria/HySea/UTFSM-IniciacionTsunami-20200505T151943Z-001/UTFSM-IniciacionTsunami/ejemplo_chile/Escenarios_corridos/'#os.getcwd()+'/'
carpeta='Escenarios/level01_txt/'#'resul_series_sce400/'

 

for i in range(248,350):
        
    if np.any(i <  int(10)) and np.any(i>=0):
        fold_escenario='00000'+str(i)+'/'
    if np.any(i < int(100)) and np.any(i >= int(10)):
        fold_escenario='0000'+str(i)+'/'
    if np.any(i >= int(100)):
        fold_escenario='000'+str(i)+'/'
        
    runs_level01=np.load(ruta_level01+fold_escenario+'level01-float16.npy',allow_pickle=True).item()
    
    lon_=runs_level01['lon']
    lat_=runs_level01['lat']
    deformed_bathy_=runs_level01['deformed_bathy']
    original_bathy_=runs_level01['original_bathy']
    
    longitud_=open(ruta+carpeta+'Lon_' + fold_escenario.split('/')[0]+'.txt','w')
    for j in range(len(lon_)):
        longitud_.write(str(lon_[j])+'\n')
    longitud_.close()
    
    latitud_=open(ruta+carpeta+'Lat_' + fold_escenario.split('/')[0]+'.txt','w')
    for j in range(len(lat_)):
        latitud_.write(str(lat_[j])+'\n')
    latitud_.close()
    
    def_bathy_=open(ruta+carpeta+'DefBathy_' + fold_escenario.split('/')[0]+'.txt','w')
    for j in range(deformed_bathy_.shape[1]):
        for k in range(deformed_bathy_.shape[0]):
            def_bathy_.write(str(deformed_bathy_[k][j])+'\t')
        def_bathy_.write('\n')
    def_bathy_.close()
    
    or_bathy_=open(ruta+carpeta+'OriginalBathy_' + fold_escenario.split('/')[0]+'.txt','w')
    for j in range(original_bathy_.shape[1]):
        for k in range(original_bathy_.shape[0]):
            or_bathy_.write(str(original_bathy_[k][j])+'\t')
        or_bathy_.write('\n')
    or_bathy_.close()
    
    print(i)
```
